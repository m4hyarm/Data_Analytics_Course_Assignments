<div dir="rtl" lang="fa" style='font-size:12.0pt;color:purple'>- با دو بار انجام عمل جمع روی مقادیر notnull ستونها و مجموع آنها تعداد کلمات به دست می آید :</div>

**<div dir="rtl" lang="fa" style='font-size:13.0pt'>در صورت امکان تعداد کلمات منحصر به فرد را گزارش کنید.</div>**

<div dir="rtl" lang="fa" style='font-size:12.0pt;color:purple'>- چون لغات استخراج شده در مرحله قبل دارای موارد اضافه مثل ،  .  ' > بود و ما در این مرحله باید لغات تکراری را حذف کنیم پس با استفاده از کتابخانه Regular Expressions فقط لغات را با پترن '[a-zA-Z''0-9]+' از متن استخراج کرده و داخل یک ست ریختم. ست ها مقادیر تکراری را نمی پذیرند، پس الان یک ست با کلمات منحصربفرد داریم که اندازه آن تعداد کلمات را مشخص می کند :</div>

```python
import re

pattern = '[a-zA-Z''0-9’]+'

words = set()
for i in text:
    for match in re.finditer(pattern, i, re.S):
        words.add(match.group().lower())

len(words)
```

<div dir="rtl" lang="fa" style='font-size:15.0pt;color:#00B0F0'>داده های میزان ابتلا به کرونا و مرگ و میر ناشی از کرونا را به ازای یک کشور دلخواه (داده های مربوط به سال 2021) دانلود کنید و به سوالات زیر پاسخ دهید :</div>

<div dir="rtl" lang="fa" style='font-size:12.0pt;color:purple'>- داده های میزان ابتلای جهان را از <a href="https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/jhu/full_data.csv">این لینک</a> دانلود کرده و داده های خواسته شده مسئله را برای کشور ایتالیا و سال 2021 داخل جدول ریختم و index جدول را ستون تاریخ قرار دادم :</div>

```python
covid = pd.read_csv('jhu-covid-data.csv', index_col='date')
IT_covid = covid[covid.location == 'Italy']
IT_covid_2021 = IT_covid.loc['2021':,'new_cases':'total_deaths']
IT_covid_2021.head()
```

<div dir="rtl" lang="fa" style='font-size:12.0pt;color:purple'>- همچنین داده های میزان واکسیناسیون جهان را از <a href="https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/vaccinations/vaccinations.csv">این لینک</a> دانلود کرده و داده های خواسته شده مسئله را برای کشور ایتالیا و سال 2021 داخل جدول ریختم و index جدول را ستون تاریخ قرار دادم :</div>

```python
vaccines = pd.read_csv('owid-vaccinations-data.csv', index_col='date')
IT_vaccines = vaccines[vaccines.location == 'Italy']
IT_vaccines_2021 = IT_vaccines.loc['2021':,['total_vaccinations', 'daily_vaccinations']]
IT_vaccines_2021.head()
```

**<div dir="rtl" lang="fa" style='font-size:13.0pt'>داده های دو جدول واکسن و ابتلا به کرونا را با توجه به تاریخ ادغام کنید.</div>**

<div dir="rtl" lang="fa" style='font-size:12.0pt;color:purple'>- با استفاده از دستور merge کتابخانه پانداز دو جدول به دست آمده از مرحله قبل را با توجه به ستون تاریخ ادغام کردم :</div>

```python
IT = pd.merge(IT_covid_2021, IT_vaccines_2021, on = 'date')
IT.head()
```

**<div dir="rtl" lang="fa" style='font-size:13.0pt'>به ازای یک کشور دلخواه، نمودار تزریق / ابتلا به کرونا را رسم کنید.</div>**

<div dir="rtl" lang="fa" style='font-size:12.0pt;color:purple'>- ستونی جدید به نام vaccinations_to_new_cases_Ratio که نشان دهنده نسبت واکسیناسیون به مبتلاهای جدید است را در جدول ایجاد کرده و نمودار مربوط رو برای بهتر نشان دادن تغییرات رسم کردم :</div>

```python
IT['vaccinations_to_new_cases_Ratio'] = IT['total_vaccinations']/IT['new_cases']
IT
```

```python
x = IT.index
y = IT['vaccinations_to_new_cases_Ratio']
fig, axis = plt.subplots(figsize=[15,5])
axis.plot(x,y)
start, end = axis.get_xlim()
axis.xaxis.set_ticks(np.arange(start, end, 16))
plt.title('Italy New Cases to Vaccinations Plot')
plt.xlabel('Date')
plt.ylabel('New Cases to Total Vaccinations Ratio (log)')
plt.xticks(rotation=45);
```
